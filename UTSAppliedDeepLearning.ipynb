```python
from ultralytics import YOLO
# Membuat model YOLO baru
model = YOLO('yolov8n.yaml')
```

```python
# Load model YOLO yang sebelumnya sudah dilatih oleh ulralytics yang sudah berbobot
model = YOLO('yolov8n.pt')
```

```python
# Train model dengan data coco128.yml yang disediaakan oleh ultraytics
results = model.train(data='coco128.yaml', epochs=3)
```

```python
# Evaluasi model dengan dataset validasi
results = model.val()
```

```python
# Melakukan deteksi objek dengan model yang sudah dibuat
results = model('https://ultralytics.com/images/bus.jpg')
```

```python
# inferensi model
model.predict('bus.jpg', save=True, imgsz=640)
```
